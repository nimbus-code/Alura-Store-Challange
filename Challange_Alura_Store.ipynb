##**Importación de datos**

```python
import pandas as pd
url1 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv'
url2 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv'
url3 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv'
url4 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv'

tienda1 = pd.read_csv(url1,sep=',')
tienda2 = pd.read_csv(url2,sep=',')
tienda3 = pd.read_csv(url3,sep=',')
tienda4 = pd.read_csv(url4,sep=',')

display(tienda1)
```

##**1. Análisis de facturación**

```python
#Total de ingresos por tienda

tot_ingreso1 = tienda1['Precio'].sum()
tot_ingreso2 = tienda2['Precio'].sum()
tot_ingreso3 = tienda3['Precio'].sum()
tot_ingreso4 = tienda4['Precio'].sum()

int(tot_ingreso1)

tot_ingreso1 / 1000000
```

```python
print(f'El total de ingresos de la tienda 1 fue de: {tot_ingreso1}')
print(f'El total de ingresos de la tienda 2 fue de: {tot_ingreso2}')
print(f'El total de ingresos de la tienda 3 fue de: {tot_ingreso3}')
print(f'El total de ingresos de la tienda 4 fue de: {tot_ingreso4}')
```

```python
#Los valores totales se normalizan para que sean más fáciles de leer (millones)

tot_ingresos ={
    'Tienda 1': int(tot_ingreso1/1000000),
    'Tienda 2': int(tot_ingreso2/1000000),
    'Tienda 3': int(tot_ingreso3/1000000),
    'Tienda 4': int(tot_ingreso4/1000000)
}

print(tot_ingresos)
```

##**2. Ventas por categoría**

```python
#Calcular la cantidad de productos vendidos por categoría en cada tienda, se deben de mostrar las categorías más populares de cada tienda.

ventas_cat1 = tienda1.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
ventas_cat2 = tienda2.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
ventas_cat3 = tienda3.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
ventas_cat4 = tienda4.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
```

```python
print(f'Categorías de los productos más populares en la tienda 1 son:\n')
display(ventas_cat1.head(5))
print(f'Categorías de los productos más populares en la tienda 2 son:\n')
display(ventas_cat2.head(5))
print(f'Categorías de los productos más populares en la tienda 3 son:\n')
display(ventas_cat3.head(5))
print(f'Categorías de los productos más populares en la tienda 4 son:\n')
display(ventas_cat4.head(5))
```

```python
#Ventas totales por categoría

sumventascat_1 = tienda1.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
sumventascat_2 = tienda2.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
sumventascat_3 = tienda3.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
sumventascat_4 = tienda4.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
```

```python
print('Tienda 1 : Ventas totales (ingresos) por categoría')
display(sumventascat_1)
print('-------------------------------------')
print('Tienda 2 : Ventas totales (ingresos) por categoría')
display(sumventascat_2)
print('-------------------------------------')
print('Tienda 3 : Ventas totales (ingresos) por categoría')
display(sumventascat_3)
print('-------------------------------------')
print('Tienda 4 : Ventas totales (ingresos) por categoría')
display(sumventascat_4)
```

##**3. Calificación promedio de la tienda**

```python
#Se calculan las calificaciones promedio de los clientes para cada tienda, con el objetivo de conocer la satisfacción del cliente con los productos vendidos

sat_producto_t1 = tienda1.groupby('Categoría del Producto')['Calificación'].mean().sort_values(ascending=False)
sat_producto_t2 = tienda2.groupby('Categoría del Producto')['Calificación'].mean().sort_values(ascending=False)
sat_producto_t3 = tienda3.groupby('Categoría del Producto')['Calificación'].mean().sort_values(ascending=False)
sat_producto_t4 = tienda4.groupby('Categoría del Producto')['Calificación'].mean().sort_values(ascending=False)
```

```python
#Calificación promedio por tipo de producto

print('Calificación promedio por categoria en la Tienda 1:')
display(sat_producto_t1)
print('Calificación promedio por categoria en la Tienda 2:')
display(sat_producto_t2)
print('Calificación promedio por categoria en la Tienda 3:')
display(sat_producto_t3)
print('Calificación promedio por categoria en la Tienda 4:')
display(sat_producto_t4)
```

```python
#Promedio de satisfación total por tienda

sat1 = sat_producto_t1.mean()
sat2 = sat_producto_t2.mean()
sat3 = sat_producto_t3.mean()
sat4 = sat_producto_t4.mean()
```

```python
print(f'Valoración promedio de tienda por los productos vendidos')
print(f'Tienda 1: {sat1:2f}\nTienda 2: {sat2:2f}\nTienda 3: {sat3:2f}\nTienda 4: {sat4:2f}')
```

##**4. Productos más o menos vendidos**

```python
#Identificar los productos que tuvieron más y menos ventas en cada tienda

vent_t1 = tienda1.groupby('Producto')['Producto'].count().sort_values(ascending=False)
vent_t2 = tienda2.groupby('Producto')['Producto'].count().sort_values(ascending=False)
vent_t3 = tienda3.groupby('Producto')['Producto'].count().sort_values(ascending=False)
vent_t4 = tienda4.groupby('Producto')['Producto'].count().sort_values(ascending=False)
```

```python
#Se guardan las variables con los productos que tuvieron mayores ventas

masvp1 = vent_t1.head()
masvp2 = vent_t2.head()
masvp3 = vent_t3.head()
masvp4 = vent_t4.head()
```

```python
#Se guardan las variables con los productos que tuvieron menores ventas

menosvp1 = vent_t1.tail()#.sort_values(ascending=True)
menosvp2 = vent_t2.tail()
menosvp3 = vent_t3.tail()
menosvp4 = vent_t4.tail()
```

```python
print('Tienda 1: Productos más vendidos')
display(masvp1)
print('Tienda 1: Productos menos vendidos')
display(menosvp1)
print('----------------------------------------')
print('Tienda 2: Productos más vendidos')
display(masvp2)
print('Tienda 2: Productos menos vendidos')
display(menosvp2)
print('----------------------------------------')
print('Tienda 3: Productos más vendidos')
display(masvp3)
print('Tienda 3: Productos menos vendidos')
display(menosvp3)
print('----------------------------------------')
print('Tienda 4: Productos más vendidos')
display(masvp4)
print('Tienda 4: Productos menos vendidos')
display(menosvp4)
print('----------------------------------------')
```

##**5. Envío promedio por tienda**

```python
#Se calcula el costo de envío promedio para cada tienda, para comprender cuánto se gasta (el cliente es el que paga el envío)

mean_envio_t1 = float(tienda1['Costo de envío'].mean())
mean_envio_t2 = float(tienda2['Costo de envío'].mean())
mean_envio_t3 = float(tienda3['Costo de envío'].mean())
mean_envio_t4 = float(tienda4['Costo de envío'].mean())
```

```python
print(f'El promedio de costos por envío en cada tienda fue de: \n')
print(f'Tienda 1: {mean_envio_t1:2f} \nTienda 2: {mean_envio_t2:2f} \nTienda 3: {mean_envio_t3:2f} \nTienda 4: {mean_envio_t4:2f}')
```
